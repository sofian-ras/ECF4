```python
import re
# pip install nltk
import nltk
# pip install spacy
import spacy
from nltk.corpus import stopwords

nltk.download("stopwords")
# python -m spacy download fr_core_news_sm
nlp = spacy.load("fr_core_news_sm")

texte = "Bonjour! Ce film est absolument MAGNIFIQUE! J'ai adoré chaque minute."

# 1. Minuscules
texte_lower = texte.lower()
print(texte_lower)

# 2. Supprimer la ponctuation
texte_clean = re.sub(r'[^a-z\s]', "", texte_lower)
print(texte_clean)

# 3. Tokenization
mots = texte_clean.split()
print(mots)

# 4. Supprimer les stopwords
stop_words = set(stopwords.words('french'))
mots_filtered = [m for m in mots if m not in stop_words]
print(mots_filtered)

# 5. Lemmatisation avec Spacy
doc = nlp(" ".join(mots_filtered))
lemmas = [token.lemma_ for token in doc]
print(lemmas)

print(f"Resultat final : {' '.join(lemmas)}")
```
