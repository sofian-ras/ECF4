```python
from sklearn.feature_extraction.text import CountVectorizer

# Documents
docs = [
    "python machine learning python",
    "python et donnees",
    "machine learning et donnees"
]

# Create CountVectorizer
vectorizer = CountVectorizer(lowercase=True)
X = vectorizer.fit_transform(docs)

print("Vocabulaire:", vectorizer.get_feature_names_out())
print("Matrice sparse shape:", X.shape)
print("Document 1 (dense):", X[0].toarray())
print("Document 2 (dense):", X[1].toarray())
print("Document 3 (dense):", X[2].toarray())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
from nltk.corpus import stopwords

# Documents exemple
docs = [
    "python est un langage de programmation",
    "machine learning utilise python",
    "data science avec machine learning",
    "deep learning et neural networks"
]

french_stopwords = stopwords.words('french')

# Creer et fit TfidfVectorizer
vectorizer = TfidfVectorizer(
    max_features=20,  # Garder top 20 mots
    stop_words=french_stopwords,  # Supprimer stopwords
    min_df=1,  # Mot doit appara dans min 1 doc
    max_df=0.8  # Mot ne doit pas appara dans >80% docs
)

X = vectorizer.fit_transform(docs)

print(X)

# Afficher les features (mots)
features = vectorizer.get_feature_names_out()
print(f"Vocabulaire ({len(features)} mots):", features)

# Afficher les scores TF-IDF pour le premier document
tfidf_scores = X[0].toarray()
for i, score in enumerate(tfidf_scores[0]):
    if score > 0:
        print(f"{features[i]:20} : {score:.4f}")
```

```python
# Utiliser n-grams avec TfidfVectorizer
vectorizer = TfidfVectorizer(
    ngram_range=(1, 2),  # Unigrams + bigrams
    max_features=100
)

X = vectorizer.fit_transform(docs)
features = vectorizer.get_feature_names_out()

# Afficher les features
for i, f in enumerate(features[:20]):
    print(f"{i+1:2}. {f}")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import Pipeline
import re

# Preprocessing custom
def clean_text(text):
    # Minuscules
    text = text.lower()
    # Supprimer ponctuation
    text = re.sub(r'[^a-z\s]', '', text)
    # Espaces extras
    text = re.sub(r'\s+', ' ', text).strip()
    return text

# Appliquer preprocessing avant vectorization
docs = ["Ceci est un exemple!", "Voila un autre exemple."]
docs_cleaned = [clean_text(d) for d in docs]

# Vectorize
vectorizer = TfidfVectorizer(
    max_features=100,
    ngram_range=(1, 2),
    min_df=1,
    max_df=0.9,
    stop_words=french_stopwords
)

X = vectorizer.fit_transform(docs_cleaned)
features = vectorizer.get_feature_names_out()
print(f"Shape: {X.shape}")  # (2, num_features)
print(X)
print(features)
# (document, index)
```
