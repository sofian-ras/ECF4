```python
tweets = [
    "Just finished my breakfast!! Check out https://t.co/xyz #goodMorning ☀️😊",
    "Loving the new #Python3.10 features... performance is amazing!!!",
    "@john_doe The movie was REALLY bad... waste of time :(",
    "Don't miss our special offer!!! Visit our website: https://example.com",
    "I can't believe it's Friday!!! #TGIF #Weekend",
    "Machine Learning & Deep Learning are the future of tech!!! 🚀🚀🚀"
]
```

### Tâche 1 : Conversion et gestion de base 

```python
def preprocess_tweet(tweet):
    # convertir en minuscules
    tweets = tweet.lower()

    # supprimer les URLs
    tweets = re.sub(r'http\S+|www\S+', "", tweets)

    # supprimer les mentions et hashtags
    tweets = re.sub(r'@\w+', "", tweets)
    tweets = re.sub(r'#\w+', "", tweets)

    # supprimer les emojis
    tweets = emoji.replace_emoji(tweets, replace="")

    # remplacer les caractères dupliqués excessifs (!!!, ???, ...) par un seul exemplaire
    tweets = re.sub(r'([!?.])\1+', r'\1', tweets)

    return tweets
```

```python
clean_tweets = []

for tweet in tweets:
    clean_tweets.append(preprocess_tweet(tweet))

    print(clean_tweets)
```

### Tâche 2 : Gestion de la ponctuation et des caractères spéciaux

```python
tweet = "I can't believe ... it's AMAZING!!!"

def preprocess_tweet(tweet):
    # convertir en minuscules
    tweets = tweet.lower()

    # supprimer les URLs
    tweets = re.sub(r'http\S+|www\S+', "", tweets)

    # supprimer les mentions et hashtags
    tweets = re.sub(r'@\w+', "", tweets)
    tweets = re.sub(r'#\w+', "", tweets)

    # supprimer les emojis
    tweets = emoji.replace_emoji(tweets, replace="")

    # remplacer les caractères dupliqués excessifs (!!!, ???, ...) par un seul exemplaire
    tweets = re.sub(r'([!?.])\1+', r'\1', tweets)

    # supprimer la ponctuation inutile sauf les apostrophes
    tweets = re.sub(r'[^\w\s\']', '', tweets)

    # supprimer les espaces multiples
    tweets = re.sub(r'[\s]+', ' ', tweets).strip()

    # Gérez les cas particuliers :
    # - "can't" doit rester "can't" (contraction)
    # - "don't" doit rester "don't"

    tweets = re.sub(r'\b(can|do)n\'t\b', r'\1n\'t', tweets)

    return tweets

clean_tweet = preprocess_tweet(tweet)
print(clean_tweet)
```

### Tâche 3 : Tokenization et stopwords

```python
#tokenisez le texte nettoyé (utilisez NLTK ou une apporche simple split)
tokens = clean_tweet.split()
print(f"Tokenization: {tokens}")

# creez une liste personalisée de stopwords en francais et anglais
stopwords = set([
    "the", "is", "in", "and", "to", "of", "a", "that", "it", "with",
    "just", "my", "out", "check", "new", "features", "performance",
    "amazing", "movie", "was", "really", "bad", "waste", "time",
    "don't", "miss", "our", "special", "offer", "visit", "website",
    "can't", "believe"
])

# supprimez les stopwords
sup_stopwords_tokens = [token for token in tokens if token in stopwords]
print(f"stopwords supprimés : {sup_stopwords_tokens}")
```

### Tâche 4 : Lemmatisation

```python
#Implémentez la lemmatisation avec NLTK WordNetLemmatizer
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet

lemmatizer = WordNetLemmatizer()

# Télécharger les ressources NLTK nécessaires (à faire une seule fois)
nltk.download('averaged_perceptron_tagger_eng', quiet=True)
nltk.download('wordnet', quiet=True)
nltk.download('omw-1.4', quiet=True)

def get_wordnet_pos(word):
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}
    return tag_dict.get(tag, wordnet.NOUN)

lemmatized_tokens = [lemmatizer.lemmatize(token, get_wordnet_pos(token)) for token in sup_stopwords_tokens]
print(f"Lemmatization: {lemmatized_tokens}")

## Conseils

#- Utilisez `re` (regex) pour les suppressions complexes
#- NLTK propose `nltk.corpus.stopwords.words('english')` et `'french'`
#- `WordNetLemmatizer` nécessite le POS tagging pour être vraiment efficace
#- SpaCy est plus puissant que NLTK pour la lemmatisation
#- Testez régulièrement pour voir l'impact de chaque étape

```

### Tâche 5 : Pipeline complet

```python
def clean_tweet(text):
    """Pipeline complet de nettoyage de tweet"""
    # Étapes 1 & 2: Nettoyage de base et ponctuation
    text = text.lower()
    text = re.sub(r'http\S+|www\S+', "", text)
    text = re.sub(r'@\w+|#\w+', "", text)
    text = emoji.replace_emoji(text, replace="")
    text = re.sub(r'([!?.])\1+', r'\1', text)
    text = re.sub(r'[^\w\s\']', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    
    # Étape 3: Tokenization et suppression des stopwords
    stopwords = {"the", "is", "in", "and", "to", "of", "a", "an", "that", "it", "with",
                 "for", "on", "at", "by", "from", "as", "are", "be", "this", "or",
                 "but", "not", "have", "has", "had", "do", "does", "did", "was", "were",
                 "been", "being", "i", "you", "he", "she", "we", "they", "my", "your",
                 "his", "her", "its", "our", "their"}
    
    tokens = [token for token in text.split() if token not in stopwords]
    
    # Étape 4: Lemmatisation
    return [lemmatizer.lemmatize(token, get_wordnet_pos(token)) for token in tokens]

# Application sur le corpus
clean_tweets_final = [clean_tweet(tweet) for tweet in tweets]
print(f"Pipeline terminé! {len(clean_tweets_final)} tweets nettoyés")
print(clean_tweets_final)
```

### Bonus 2 : Gestion des cas spéciaux français

```python
textes_francais = [
    "C'est l'été à Paris! J'adore cette saison ☀️",
    "Rendez-vous après-midi pour un café crème",
    "L'intelligence artificielle révolutionne l'économie française"
]

def clean_french(text):
    text = text.lower()
    text = emoji.replace_emoji(text, replace="")
    text = re.sub(r'\b([lcdjmnts])\'\s*', r'\1 ', text)  # élisions: l' → l
    text = re.sub(r'-', ' ', text)  # traits d'union
    text = re.sub(r'[^\w\sàâäéèêëïîôùûüÿæœç]', '', text)
    return re.sub(r'\s+', ' ', text).strip()

print("GESTION DES CAS FRANÇAIS")
print("---------------")
for texte in textes_francais:
    print(f"\n→ {texte}")
    print(f"  {clean_french(texte)}")
```

### Bonus 3 : Classe TextCleaner réutilisable

```python
class TextCleaner:
    """Classe réutilisable pour le nettoyage de texte"""
    
    def __init__(self, use_lemmatization=True, use_stemming=False, custom_stopwords=None):
        self.use_lemmatization = use_lemmatization
        self.use_stemming = use_stemming
        self.stopwords = custom_stopwords or {
            "the", "is", "in", "and", "to", "of", "a", "an", "that", "it", "with",
            "for", "on", "at", "by", "from", "as", "are", "be", "this", "or",
            "but", "not", "have", "has", "had", "do", "does", "did", "was", "were",
            "been", "being", "i", "you", "he", "she", "we", "they", "my", "your"}
        
        if use_lemmatization:
            self.lemmatizer = WordNetLemmatizer()
        if use_stemming:
            self.stemmer = PorterStemmer()
    
    def clean(self, text):
        """Nettoyer un texte"""
        text = text.lower()
        text = re.sub(r'http\S+|www\S+|@\w+|#\w+', "", text)
        text = emoji.replace_emoji(text, replace="")
        text = re.sub(r'[^\w\s\']', '', text)
        text = re.sub(r'\s+', ' ', text).strip()
        
        tokens = [t for t in text.split() if t not in self.stopwords]
        
        if self.use_lemmatization:
            return [self.lemmatizer.lemmatize(t, get_wordnet_pos(t)) for t in tokens]
        elif self.use_stemming:
            return [self.stemmer.stem(t) for t in tokens]
        return tokens
    
    def clean_corpus(self, texts):
        """Nettoyer un corpus de textes"""
        return [self.clean(text) for text in texts]
```

```python
# Tests de la classe TextCleaner
print("TESTS TEXTCLEANER")
print("-------")

configs = [
    ("Lemmatisation", TextCleaner(use_lemmatization=True, use_stemming=False)),
    ("Stemming", TextCleaner(use_lemmatization=False, use_stemming=True)),
    ("Sans traitement", TextCleaner(use_lemmatization=False, use_stemming=False))
]

for nom, cleaner in configs:
    print(f"\n{nom}: {' '.join(cleaner.clean(tweets[0]))}")
```
