## Callbacks

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, callbacks
import numpy as np
import matplotlib.pyplot as plt
```

```python
(X_train, y_train), (X_test, y_test) = keras.datasets.cifar10.load_data()

X_train_norm = X_train.astype("float32") / 255.0
X_test_norm = X_test.astype("float32") / 255.0

val_size = 5000
X_val = X_train_norm[-val_size:]
y_val = y_train[-val_size:]
X_train = X_train_norm[:-val_size]
y_train = y_train[:-val_size]

print(X_train.shape)
print(X_val.shape)
print(X_test.shape)
```

```python
def create_cnn_model():
    model = keras.Sequential([
        layers.Conv2D(32, 3, activation="relu", input_shape=(32,32,3)),
        layers.BatchNormalization(),
        layers.MaxPooling2D(2),
        layers.Conv2D(64, 3, activation="relu"),
        layers.BatchNormalization(),
        layers.MaxPooling2D(2),

        layers.Flatten(),
        layers.Dense(128, activation="relu"),
        layers.Dense(10, activation="softmax")
    ])

    return model

model_no_callbacks = create_cnn_model()
model_no_callbacks.summary()
```

```python
model_no_callbacks.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

history_no_callbacks = model_no_callbacks.fit(
    X_train, y_train,
    batch_size=128,
    epochs=10,
    validation_data=(X_val, y_val),
    verbose=1
)
```

```python
import os

os.makedirs("models", exist_ok=True)

callback_list = [
    # EarlyStopping : arrГӘte si val_loss ne s'amГ©liore plus
    callbacks.EarlyStopping(
        monitor="val_loss",
        patience=3,
        restore_best_weights=True,
        verbose=1
    ),
    # ReduceLROnPlateau : RГ©duit le LR si val_loss stagne
    callbacks.ReduceLROnPlateau(
        monitor="val_loss",
        patience=3,
        factor=0.5,
        verbose=1
    ),
    # ModelCheckpoint : sauvegarde le meilleur model
    callbacks.ModelCheckpoint(
        filepath="models/best_model_demo.h5",
        monitor="val_accuracy",
        save_best_only=True,
        mode="max",
        verbose=1
    )
]
```

```python
model_with_callback = create_cnn_model()

model_with_callback.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

history_with_callback = model_with_callback.fit(
    X_train, y_train,
    batch_size=128,
    epochs=10,
    validation_data=(X_val, y_val),
    callbacks=callback_list,
    verbose=1
)
```
